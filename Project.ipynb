## Data Preparation and Cleaning

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

````python
india_crime_data=pd.read_csv('District_wise_crimes_committed_IPC_2001_2012.csv')
````

This is the whole india crime data(india_crime_data) i wanted to analysis only on bihar data so we need to extract bihar data from india_data,
here are the beggining of my data cleaning

````python
india_crime_data
````

I am using (.loc) method here for extraction of row which is starting with BIHAR

````python
bihar_crime_data=india_crime_data.loc[india_crime_data.STATE_UT=="BIHAR"]
````

In this data we found 'TOTAL' in district column which we dont want in our data, so we need to drop the row from our data

````python
bihar_crime_data
````

#### bihar_crime_data1 is a copy of bihar_crime_data for future reference

````python
bihar_crime_data1=bihar_crime_data.copy()
````

#### here i am setting "DISTRICT"  as index because we do not want to "TOTAL" in our DISTRICT column

````python
bihar_crime_data=bihar_crime_data.set_index("DISTRICT")
````

````python
bihar_crime_data
````

#### Here drop method is using for deleting the total column from row

````python
bihar_crime_data=bihar_crime_data.drop('TOTAL',axis=0)
````

#### This is my final bihar crime data

````python
bihar_crime_data
````

#### There is no use of STATE_UT Column so i am deleting it 

````python
bihar_crime_data=bihar_crime_data.drop('STATE_UT',axis=1)
bihar_crime_data
````

#### below the heatmap showing Murder in every districts of bihar from 2001 to 2012
#### after analysis of this map we easily identiffy the most violent district of bihar which is patna

````python
murders=bihar_crime_data2.pivot("YEAR","DISTRICT","MURDER")
f,ax = plt.subplots(figsize=(30, 7));
sns.heatmap(murders,annot=True, fmt="d", linewidths=.5, ax=ax);
plt.title(" Murder in districts of bihar(2001-2012)");
````

**The data set below is showing sum of all the particular crime in all the district of bihar year wise, this data set tell us murder in all 44 district in 2001 is 3643**

````python
year_wise_data=bihar_crime_data2.groupby("YEAR").sum()
year_wise_data
````

**Below i have plotted a bar grap which is showing rape in bihar yearly. after seeing this graph we can easily say that the maximum rape was in 2007**

````python
plt.figure(figsize=(20,7));
plt.xticks(rotation=75);
plt.title("Rapes in bihar(2001-2012)");
sns.barplot(x=year_wise_data.index,y=year_wise_data.RAPE,);
````

**below i have plotted a line graph which is showing how kidnaping increasing year by year, this is very dangeorus**

````python
kindnapping_yearly=year_wise_data.sort_values('KIDNAPPING',ascending=False)
````

````python
kindnapping_yearly
````

````python
with plt.style.context('Solarize_Light2'):
            plt.title("Kidnapping in bihar(2001-2012)");
            plt.xlabel("YEAR");
            plt.ylabel("Kidnapping")
            plt.plot(kindnapping_yearly.KIDNAPPING);
````

**Below i have plotted a subplot in which i try to show various serious crime are growing year by year in bihar** 

````python
fig, axes = plt.subplots(2, 3, figsize=(25, 12));

axes[0,0].set_title("Dacoity cases in BIHAR(2001-2012)");
plt.xlabel('Year');
plt.ylabel('Dacoity');
axes[0,0].bar(year_wise_data.index, year_wise_data.DACOITY, color = 'black');


axes[0,1].set_title("Robbery cases in BIHAR(2001-2012)");
plt.xlabel('Year');
plt.ylabel('Robbery');
axes[0,1].bar(year_wise_data.index, year_wise_data.ROBBERY, color = 'blue');


axes[0,2].set_title("Burglary cases in BIHAR(2001-2012)");
plt.xlabel('Year');
plt.ylabel('Burglary') ;
axes[0,2].bar(year_wise_data.index, year_wise_data.BURGLARY, color = 'yellow');



axes[1,0].set_title("Theft cases in BIHAR(2001-2012)");
plt.xlabel('Year');
plt.ylabel('Theft');
axes[1,0].bar(year_wise_data.index, year_wise_data.THEFT, color = 'red');


axes[1,1].set_title("Dowry deaths in BIHAR(2001-2012)");
plt.xlabel('Year');
plt.ylabel('Dowry deaths');
axes[1,1].bar(year_wise_data.index, year_wise_data.DOWRY_DEATHS, color = 'orange');

axes[1,2].set_title("cruelty by husband in BIHAR(2001-2012)");
plt.xlabel('Year');
plt.ylabel('cruelty by husband');
axes[1,2].bar(year_wise_data.index, year_wise_data.CRUELTY_BY_HUSBAND_OR_HIS_RELATIVES, color = 'green');
````

````python
year_wise_data
````

````python
yearly_ipc_crime=year_wise_data.sort_values('TOTAL_IPC_CRIMES',ascending=False)
yearly_ipc_crime
````

**below i have plotted a horizontal bar graph which is showing IPC crime record(FIR IN POLICE STATION) year wise**

````python
fig, ax = plt.subplots();
plt.title("Yearly IPC crime Record")
plt.xlabel("IPC case")
plt.ylabel("YEAR")
ax.barh(yearly_ipc_crime.index,yearly_ipc_crime.TOTAL_IPC_CRIMES , height=0.7, left=None,align='center');
````

#### Q5: percentage of murder in top 5 districts in comparison of all murder in all districts?

````python
district_wise_crime=bihar_crime_data.groupby('DISTRICT').sum().sort_values("MURDER",ascending=False)
````

````python
district_wise_crime
````

````python
top_five_distict_total_murder=district_wise_crime.head(5).MURDER.sum()
````

````python
total_murder_in_all_districts=district_wise_crime.MURDER.sum()
````

````python
percentage_of_murder_in_top_5_districts=(top_five_distict_total_murder/total_murder_in_all_districts)*100
````

````python
print(" percentage of murder in top 5 districts is {:.2f} %".format(percentage_of_murder_in_top_5_districts))
````
